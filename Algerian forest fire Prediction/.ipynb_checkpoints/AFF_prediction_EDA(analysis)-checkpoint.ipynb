FWI Components  
6. Fine Fuel Moisture Code (FFMC) index from the FWI system: 28.6 to 92.5 
7. Duff Moisture Code (DMC) index from the FWI system: 1.1 to 65.9 
8. Drought Code (DC) index from the FWI system:  7 to 220.4
9. Initial Spread Index (ISI) index from the FWI system: 0 to 18.5 
10. Buildup Index (BUI) index from the FWI system: 1.1 to 68
11. Fire Weather Index (FWI) Index: 0 to 31.1
12. Classes: two classes, namely   Fire and not Fire

The correlation matrix reveals several notable relationships among the variables. Relative Humidity (RH) has a strong negative correlation with these same fire risk indicators, suggesting that lower humidity levels are linked to higher fire risks. Wind speed (Ws) shows a weak correlation with other variables, indicating it has a lesser impact on fire risk. Rainfall negatively correlates with fire risk indicators, showing that increased rainfall reduces fire risk. The variable 'Classes' is strongly correlated with FFMC, FWI, and ISI, highlighting their importance in classifying fire events. The regional variable shows weaker correlations, suggesting that other factors have a more direct influence on fire risk.
